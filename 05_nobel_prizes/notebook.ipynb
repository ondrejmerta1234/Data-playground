The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Start coding here!
```

```python
nobel = pd.read_csv("data/nobel.csv")
print(nobel.columns)
gender_count = nobel.groupby("sex")["laureate_id"].count()
print(gender_count)
print()
top_gender = gender_count.idxmax()
print(top_gender)

country_count = nobel.groupby("birth_country")["laureate_id"].count()
print(country_count)
top_country = country_count.idxmax()
print(top_country)
```

```python
us_born = nobel.query("birth_country == 'United States of America'")
print(us_born)
us_born['year_decade'] = us_born['year'] // 10
us_born_per_dec = us_born.groupby("year_decade")["laureate_id"].count()
print(us_born_per_dec)

nobel['year_decade'] = nobel['year'] // 10
nobel_per_dec = nobel.groupby("year_decade")["laureate_id"].count()
print(nobel_per_dec)

res = us_born_per_dec / nobel_per_dec
print(res)
max_decade_usa = res.idxmax() * 10
print(max_decade_usa)

```

```python
nobel_women = nobel.query('sex == "Female"')
fem_per_dec = nobel_women.groupby(["year_decade", "category"])["laureate_id"].count()
all_per_dec = nobel.groupby(["year_decade", "category"])["laureate_id"].count()

#print(fem_per_dec)
#print(all_per_dec)

portion = fem_per_dec / all_per_dec
portion = portion.sort_values(ascending=False)
print(portion)
res = portion.head(1)
res = res.index
decade, category = res[0]
decade = int(decade * 10)
max_female_dict = {decade:category}
print(max_female_dict)


```

```python
nobel_women = nobel.query('sex == "Female"')
print(nobel_women)
index = nobel_women["year"].idxmin()
first_woman = nobel_women.loc[index]
print(first_woman)
first_woman_name = first_woman["full_name"]
print(first_woman_name)
first_woman_category = first_woman["category"]
print(first_woman_category)
```
