Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# Start coding here...
best_math_schools = schools.sort_values("average_math",ascending=False)
best_math_schools = best_math_schools.loc[:, ["school_name", "average_math"]]
best_math_schools["percentage"] = (best_math_schools["average_math"] / 800) * 100
best_math_schools = best_math_schools.query("percentage >= 80")
print(best_math_schools.loc[:,["percentage", "average_math"]])
best_math_schools = best_math_schools.loc[:, ["school_name", "average_math"]]

# Add as many cells as you like...
```

```python
top_10_schools = schools.copy() #we will work on a schools copy
melted = top_10_schools.melt(id_vars=['school_name', 'borough', 'building_code', 'percent_tested'])
#melts for better use of groupby
print(melted.loc[:,["school_name","variable", "value"]].head())
melted = melted.loc[:, ['school_name', "variable", "value"]]


total = melted.groupby("school_name").agg(total_SAT=("value", "sum")).reset_index()
#groups by school_name and sums score from all tests

top_10_schools = total
top_10_schools = top_10_schools.sort_values("total_SAT", ascending=False)
top_10_schools = top_10_schools.head(10) #top 10 schools
print(top_10_schools.head())



```

```python
#largest standart deviatin in borough
school_cpy = schools.merge(total, on="school_name")
school_cpy = school_cpy.groupby("borough").agg(num_schools=("school_name", "count"), average_SAT=("total_SAT","mean"), std_SAT=("total_SAT", "std"))
largest_std_dev = school_cpy.sort_values("std_SAT", ascending=False)
largest_std_dev = largest_std_dev.head(1).round(2)
print(largest_std_dev)
```
